### df.loc[row,col]
- 使用行标签与列标签对数据进行查询
- 布尔索引可参考直接切片，可以同时对行与列进行筛选
- 索引器支持列表、切片、布尔数组、多级索引、日期范围
- 与直接切片不同，对`columns`的切片，多列了`col_X:col_Y`这样的方法

````python
from pandas import read_csv
df=read_csv(
    filepath_or_buffer=r'..\0-data\beijing_tianqi\beijing_tianqi_2017-2019.csv',
    sep=',',
    usecols=range(9),
    #nrows=10,
    skiprows=1,
    header=None,
    names=['日期','最高温度','最低温度','天气','风向','风速','空气指数','空气质量','空气质量等级'],
    dtype={'空气指数':'int8','空气质量等级':'int8'},
    parse_dates=['日期'],
    date_format='%Y-%m-%d',
    encoding='utf-8',
    index_col='日期',
    converters={'最高温度':lambda x:int(x[:-1]),'最低温度':lambda x:int(x[:-1])}
).resample('MS').first()
df.head()
````

#### 日期范围
注意：
- 日期范围必须包含在索引中
- 日期左右皆包含

示例：
```python
# 获取2018年1月1日数据
df.loc['2018-01-01']        #返回Series
df.loc[['2018-01-01']]      #返回DataFrame

# 获取多个日期
df.loc[['2018-01-01', '2018-04-01']]

# 获取2018年数据
df.loc['2018']              

# 获取2018年第二季度数据
df.loc['2018-Q2']           

# 获取2018年2月到4月数据
df.loc['2018-02':'2018-04'] 

# 获取2019年前的数据
df.loc[:'2018']             
```
