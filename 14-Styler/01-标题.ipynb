```python
import pandas as pd

df = pd.read_csv(
    filepath_or_buffer=r'..\0-data\beijing_tianqi\beijing_tianqi_2018.csv',
    sep=',',
    usecols=range(8),
    nrows=20,
    skiprows=1,
    header=None,
    names=[
        '日期','最高温度','最低温度',
        '天气','风向','风速',
        '空气指数','空气质量','空气质量等级',
    ],
    dtype={'空气指数': 'int8', '空气质量等级': 'int8'},
    parse_dates=['日期'],
    date_format='%Y-%m-%d',
    encoding='utf-8',
    index_col='日期',
    converters={'最高温度': lambda x: int(x[:-1]), '最低温度': lambda x: int(x[:-1])},
)
```

标题样式无法应用于xlsx文件中  

```python
style = df.style.set_caption('北京2018年天气')  # 设置表格标题文本
caption = [  # 设置标题样式
    ('font-family', '等线'),  # 字体
    ('font-size', '24px'),  # 字号
    ('font-weight', 'bold'),  # 加粗
    ('color', 'rgb(70,130,180)'),  # 字体颜色
    ('text-align', 'center'),  # 文本对齐方式
    ('text-decoration', 'underline overline'),  # 线样式
    ('line-height', '60px'),  # 行间距
    ('height', '120px'),  # 高度
    ('width', '25vw'),  # 宽度
    ('text-align', 'end'),  # 文本对齐方式
    ('background-color', '#F0F8FF'),  # 背景颜色
    ('border', '3px dotted blue'),  # 设置边框样式
]

style = style.set_table_styles(
    [
        {
            'selector': 'caption',  # 设置标题样式
            'props': caption,
        }
    ]
)

style.to_html('01-标题.html')
style
```
