## melt
- 说明：  
    - 将DataFrame从宽格式转换为长格式。
    - 根据`id_vars`参数，指定不需要转换的列。
    - 根据`value_vars`参数，指定需要转换的列。
    
- 参数： 
    - `id_vars`: scalar | list | None  
    保持不变的列
        - None: 不考虑这个参数 (默认)
        - scalar: 列名
        - list: 列名列表
    - `value_vars`: scalar | list | None  
    需要转换的列
        - None: 使用除`id_vars`之外的所有列 (默认)
        - scalar: 列名
        - list: 列名列表
    - `var_name`: str | None  
    变量列的名称
        - None: 使用默认名称`variable` (默认)
        - str: 自定义名称
    - `value_name`: str  
    值列的名称
        - 默认值为`value`
    - `col_level`: int | str | None  
    如果列是多级索引，指定要使用的级别
        - None: 不考虑这个参数 (默认)
        - int: 索引级别
        - str: 索引级别名称
    - `ignore_index`: bool  
    是否重置索引
        - True: 重置索引 (默认)
        - False: 保留原索引

```python
import pandas as pd

df = pd.read_excel(
    r'..\0-data\stocks\互联网公司股票-02.xlsx',
    usecols='A:D'
)
df.head(10)
```

```python
df.melt(id_vars=['日期', '公司'], var_name='指标', value_name='数值')
```
