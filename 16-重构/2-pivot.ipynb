## pivot
- 说明：  
    - 将DataFrame从长格式转换为宽格式。
    - 根据`index`参数，指定用于重新索引的列。
    - 根据`columns`参数，指定用于生成新列的列。
    - 根据`values`参数，指定用于填充新DataFrame的列。
    - `pandas.pivot`和`DataFrame.pivot`功能一致,区别在于`data`参数与引用对象不同。
- 参数： 
    - `data`: DataFrame  
    用于填充新DataFrame的数据
    - `index`: scalar | list  
    用于重新索引的列
    - `columns`: scalar | list  
    用于生成新列的列
    - `values`: scalar | list | None  
    用于填充新DataFrame的列
        - None: 使用所有剩余的列 (默认)
        

    
|引用对象\参数|`data`|`index`|`columns`|`values`|
|:-------:|:-------:|:-------:|:-------:|:-------:|
|`DataFrame`||√|√|√
|`pandas`|√|√|√|√|


```python
import pandas as pd

df = pd.read_excel(r'..\0-data\stocks\互联网公司股票-03.xlsx')
df.head(10)
```

```python
pivot1 = df.pivot(index=['公司', '日期'], columns='指标', values='数值')
pivot2 = pd.pivot(df, index=['公司', '日期'], columns='指标', values='数值')
(pivot2 == pivot1).all().all()
```
