```python
import pandas as pd
df=pd.read_csv(
    filepath_or_buffer=r'..\0-data\beijing_tianqi\beijing_tianqi_2017-2019.csv',
    sep=',',
    skiprows=1,
    header=None,
    names=['日期','最高温度','最低温度','天气','风向','风速','空气指数','空气质量','空气质量等级'],
    dtype={'空气指数':'int8','空气质量等级':'int8'},
    parse_dates=['日期'],
    date_format='%Y-%m-%d',
    encoding='utf-8',
    index_col='日期',
    converters={'最高温度':lambda x:int(x[:-1]),'最低温度':lambda x:int(x[:-1])}
)
df
```

# 布尔索引
如果条件表达式的结果为True，则保留该行或列；如果结果为False，则不保留。  
根据布尔索引对应的index或columns选择行或列。  
如布尔索引为布尔索引组成的DataFrame,则保留True内容，Fales转换为空值。  
使用多个条件时，使用and（&）或or（|）运算符。 较为复杂的条件，推荐使用`query`方法。

```python
df[df.index>='2018']
```

```python
df.loc[:,df.dtypes=='int64']
```

```python
df[df.map(lambda x: isinstance(x, str))]
```

```python
df.风向[df.风向=='南风']
```
