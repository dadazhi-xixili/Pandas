### df.iloc[row,col]
- 只接受整数位置索引
- 不接受`index`或`columns`标签
- 可以接受单个整数、整数列表、整数切片
- 可以接受多个参数，第一个参数为行索引，第二个参数为列索引
- 切片方法参考list 切片方法，可以`负数指为倒数，左开右闭`

```python
from pandas import read_csv
df=read_csv(
    filepath_or_buffer=r'..\0-data\beijing_tianqi\beijing_tianqi_2017-2019.csv',
    sep=',',
    usecols=range(9),
    #nrows=10,
    skiprows=1,
    header=None,
    names=['日期','最高温度','最低温度','天气','风向','风速','空气指数','空气质量','空气质量等级'],
    dtype={'空气指数':'int8','空气质量等级':'int8'},
    parse_dates=['日期'],
    date_format='%Y-%m-%d',
    encoding='utf-8',
    index_col='日期',
    converters={'最高温度':lambda x:int(x[:-1]),'最低温度':lambda x:int(x[:-1])}
).resample('W').first()
df
```
