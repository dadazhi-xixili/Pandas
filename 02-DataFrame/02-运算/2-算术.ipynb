# 算术方法

| 方法 | 说明 |
| :--- | :--- |
| `add` | 加法 |
| `sub` | 减法 |
| `mul`| 乘法|
| `div`/`truediv`| 浮点除法|
| `floordiv`| 整除|
| `mod`| 取余|
| `pow`| 幂|
| `radd`| `反向`加法|
| `rsub`| `反向`减法|
| `rmul`| `反向`乘法|
| `rpow`| `反向`幂|
| `rmod`| `反向`取余|
| `rdiv`/`rtruediv`| `反向`浮点除法|
| `rfloordiv`| `反向`整除|
#### 参数说明:
- `other` : 操作对象
    - 任何单个或多个元素的数据结构，或类似列表的对象。
- `axis` : 轴方向
    - `0` : 行方向
    - `1` : 列方向
- `level` : 多级索引
    - `0` : 第一级索引
    - `1` : 第二级索引  
    ...
- `fill_value` : 填充值

# 其他
- `abs` 
    绝对值
    - 无参数
- `round`
    四舍五入
    - decimals : -> int|dict
        int:
            保留小数位
        dict:
            键为列名,值为保留小数位
