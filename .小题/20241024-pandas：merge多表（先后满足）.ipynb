```python
import pandas as pd

path = '20241024-pandas：merge多表（先后满足）.xlsx'
df = pd.read_excel(path, usecols='A:B', dtype={'产品名称': 'str', '总保费(元)': 'int'})
df_bf = pd.read_excel(path, usecols='E:F', nrows=4, names=['保费', '系数'])
df_cp = pd.read_excel(path, usecols='E:F', nrows=3, skiprows=6)
df.head()
```

```python
df_bf
```

```python
df_cp
```

```python
df_m1 = df.merge(df_cp,on='产品名称',how='left')
df_m1.update(df.merge(df_bf.系数,left_on='总保费(元)',right_on=df_bf.保费,how='left'))
df_m1
```

```python
df_m = df.merge(df_bf.系数,left_on='总保费(元)',right_on=df_bf.保费,how='left').fillna(df.merge(df_cp,on='产品名称'))
df_m
```
