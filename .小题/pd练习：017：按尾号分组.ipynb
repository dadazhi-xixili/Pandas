```python
import pandas as pd

df = pd.read_excel('pd练习：017：按尾号分组.xlsx', usecols='A', nrows=21)
def fx(x):
    x = x[:-1]
    s = pd.Series(x)
    i = s.str[-1]
    s.index = i
    return s

g = pd.concat(map(fx, df.数据.str.split(','))).groupby(level=0)
pd.concat(
    [
        g.get_group(x).reset_index(drop=True).rename(x)
        for x in g.groups.keys()
    ],
    axis=1,
).fillna('')
```

```python
(
    df.数据.str.split(',')
    .map(lambda x: x[:-1])
    .explode()
    .to_frame()
    .assign(c=lambda x: x.数据.str[-1])
    .reset_index(drop=True)
    .pivot(columns='c', values='数据')
    .apply(lambda x: x.dropna().reset_index(drop=True))
).fillna('')
```

```python
df = pd.read_excel('pd练习：017：按尾号分组.xlsx', usecols='A', nrows=21)
df = pd.DataFrame([','.join(df.数据)])[0].str.split(',', expand=True).T
df.index = df[0].str[-1]
g = df.groupby(df.index)
col = g.groups.keys()
c = pd.concat([g.get_group(x).reset_index(drop=1) for x in col], axis=1)
c.columns = col
c.fillna('')
```
