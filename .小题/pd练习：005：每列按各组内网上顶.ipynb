```python
import pandas as pd
df=pd.read_excel(
    'pd练习：005：每列按各组内网上顶.xlsx',
    header=0,
    usecols='A:F'
)
df
```

```python
df.groupby('日期').apply(
    lambda x: x.apply(lambda y: y.dropna(ignore_index=True)),
    include_groups=False,
).droplevel(1).fillna('')
```

```python
df.groupby(
    df.日期.copy(),
    group_keys=False,
).apply(
    lambda x: pd.concat(
        [s.dropna(ignore_index=True) for i, s in x.items()], axis=1
    ),
    include_groups=False,
).fillna('')
```
