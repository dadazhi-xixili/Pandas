```python
import pandas as pd
from cn2an import cn2an,an2cn

df = pd.read_excel('pd练习：019：对应的季度求和.xlsx', usecols='A:J', nrows=3)
df
```

```python
(
    df.melt(id_vars='姓名', var_name='季度')
    .assign(季度=lambda x: x.季度.apply(lambda y: an2cn(cn2an(y[:-1]) // 3 + 1)) + '季度')
    .pivot_table('value', '姓名', '季度', 'sum')
)
```

```python
(
    df.
    set_index('姓名').
    set_axis(
        map(lambda y: an2cn(cn2an(y[:-1]) // 3 + 1) + '季度',df.columns[1:]),
        axis=1
    ).T
    .groupby(level=0,)
    .sum().T
)
```
