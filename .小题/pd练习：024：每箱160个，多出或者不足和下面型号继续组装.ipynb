```python
import pandas as pd

df = pd.read_excel(
    'pd练习：024：每箱160个，多出或者不足和下面型号继续组装.xlsx',
    usecols='A:B',
    nrows=6,
)
df
```

```python
def f(ser: pd.Series):
    y = n - ser.cum_s % n
    l = [y if y < ser.数量 else ser.数量]
    m = ser.cum - ser.cum_s - l[0]
    while m > n:
        l.append(n)
        m -= n
    else:
        l.append(m) if m else None
    return l


n = 160
df['cum'] = df.数量.cumsum()
df['cum_s'] = df.cum.shift(1, fill_value=0)
df['箱'] = df.apply(f, axis=1)
df = df.explode('箱')
df['箱号'] =  df.箱.cumsum() // 160 +1
df
```
