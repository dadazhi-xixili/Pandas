```python
import pandas as pd

df = pd.read_excel(
    'pd练习：025：户主信息变换格式.xlsx',
    skiprows=1,
    usecols='A:G',
    nrows=17,
)
df.head()
```

```python
df.index = df.户主.notna().cumsum().values
df2 = (
    df.loc[:, '户主':'户主身份证号码']
    .dropna()
    .melt(
        id_vars='户主身份证号码',
        ignore_index=False,
    )
    .set_axis(['身份证号码', '与户主关系', '家庭成员姓名'], axis=1)
)
pd.concat(
    [df, df2],
).ffill().groupby(
    level=0, group_keys=False
).apply(lambda x: x.drop(['户主', '户主身份证号码'], axis=1))
```
