```python
import pandas as pd

df = pd.read_excel('pd练习：009：分类汇总.xlsx', usecols='A:E')
df.head()
```

```python
df.groupby('类目', group_keys=False, as_index=False).get_group('鼠标').groupby(
    '品牌'
).数量.agg(['count', lambda z: z.astype('str').str.cat(sep='+'), 'sum']).set_axis(
    ['计数', '品牌', '求和'], axis=1
).agg(
    {'计数': 'sum', '品牌': lambda z: z.str.cat(sep='+'), '求和': 'sum'}
).rename('小计')
```

```python
a = df.pivot_table(
    index=['类目', '品牌'],
    values='数量',
    aggfunc=['count', lambda x: '+'.join(x.astype('str')), 'sum'],
).set_axis(['计数', '明细', '求和'], axis=1)
b = (
    df.groupby('类目')['数量']
    .agg(计数='count', 求和='sum')
    .assign(品牌=' 小计')
    .set_index('品牌', append=True)
)
c = pd.concat([a, b])
c.groupby(level=0).apply(lambda x: x).reset_index(level=1, drop=True)
```

```python
a = df.groupby(['类目', '品牌'])['数量'].agg(
    计数='count', 明细=lambda x: '+'.join(x.astype('str')), 求和='sum'
)
b = (
    df.groupby('类目')['数量']
    .agg(计数='count', 求和='sum')
    .assign(品牌='小计')
    .set_index('品牌', append=True)
)
c = pd.concat([a, b])
c.groupby(level=0).apply(lambda x: x).reset_index(level=1, drop=True)
```
