```python
import pandas as pd
import numpy as np

df = pd.read_excel(
    '出入库存单价.xlsx',
    usecols='A:C'
)
df
```

```python
(
    df.assign(数量=lambda x: np.where(x.桩体.str[0] == '出', -x.数量, x.数量))
    .assign(结存=lambda x: x.数量.cumsum())
    .assign(结存金额=lambda x: (x.入库单价 * x.数量).cumsum() / x.结存)
    .assign(数量=lambda x: x.数量.abs())
)
```
