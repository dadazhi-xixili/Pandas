|方法|说明|参数|
|-|-|-|
|`head()`|显示前几行|默认5:`int`|
|`tail()`|显示最后几行|默认5:`int`|
|`shape`|显示行数和列数|
|`columns`|显示列名|
|`index`|显示索引|
|`dtypes`|显示列类型|
|`info()`|打印描述信息|
|`describe()`|显示统计信息|
|`size`|显示元素个数|
|`ndim`|显示维度|
|`memory_usage()`|显示内存占用情况|


```python
from pandas import read_csv
df=read_csv(
    filepath_or_buffer=r'..\0-data\beijing_tianqi\beijing_tianqi_2018.csv',
    sep=',',
    usecols=range(9),
    nrows=10,
    skiprows=1,
    header=None,
    names=['日期','最高温度','最低温度','天气','风向','风速','空气指数','空气质量','空气质量等级'],
    dtype={'空气指数':'int8','空气质量等级':'int8'},
    parse_dates=['日期'],
    date_format='%Y-%m-%d',
    encoding='utf-8',
    index_col='日期',
    converters={'最高温度':lambda x:int(x[:-1]),'最低温度':lambda x:int(x[:-1])}
)
df
```
