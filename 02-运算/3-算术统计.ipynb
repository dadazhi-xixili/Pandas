### 聚合方法
|**方法**|**描述**|`axis`|`skipna`|`min_count`|`dropna`|`numeric_only`/`bool_only`|
|:--|:--|:--:|:--:|:--:|:--:|:--:|
|`sum`|求和|1/0|bool|int|
|`avg`|平均值|1/0|bool|int||bool|
|`max`|最大值|1/0|bool|int||bool|
|`min`|最小值|1/0|bool|int||bool|
|`product`/`prod`|乘积|1/0|bool|int||bool|
|`count`|计数|1/0|bool|int||bool|
|`nunique`|去重计数|1/0|bool|int|
|`mean`|平均值|1/0|bool|int||bool|
|`mode`|众数|1/0|bool|int|bool|bool|
|`median`|中位数|1/0|bool|int||bool|
|`any`|与|1/0|bool|int||bool|
|`all`|或|1/0|bool|int||bool|

### 非聚合方法
|**方法**|**描述**|`axis`|`skipna`|`min_count`|`numeric_only`|
|:--|:--|:--:|:--:|:--:|:--:|
|`abs`|绝对值|
|`cumsum`|累计求和|√|√|√|
|`cumprod`|累计乘积|√|√|√|
|`cummax`|累计最大值|√|√|√|
|`cummin`|累计最小值|√|√|√|


### 参数:
- axis : -> int ,默认 0 
    - 轴方向
    - 0: 行
    - 1: 列
- skipna : -> bool，默认 True
    - 是否忽略缺失值
- min_count : -> int，默认 0
    - 行/列最小有效值的个数,低于此值则返回缺失值
- dropna : -> bool ,默认True 
    - 不包括空值
- numeric_only: -> bool ,默认 False
    - 仅计算包含浮点数、整数或布尔值数据轴
- bool_only: -> bool ,默认 False
    - 仅计算包含布尔值数据轴

### 其他:
- `rank` :
    - axis
    - method : str|dict ,默认 'average'
        如何对具有相同值（即相等）的记录组进行排名
        - `average`：该组的平均排名
        - `min`：组中最低排名
        - `max`：组中最高排名
        - `first`：按照数组中出现的顺序进行排序
        - `dense'`：类似“最小”，但组之间的等级总是增加 1。
    - bool_only
    - na_option : str,默认 'keep'
        如何对 NaN 值进行排序
        - `keep`： NaN 值保持不变
        - `top`： NaN 值被放置在组中的顶部
        - `bottom`： NaN 值被放置在组中的底部
    - ascending : bool ,默认 True
        排序顺序
    - pct : bool ,默认 False
        是否按百分比返回

- `value_counts` :
    - `subset` : str|list of str|slice ,默认 None
        计算唯一组合时使用的列
    - `normalize` : bool ,默认 False
        返回比例而不是频率
    - `sort` : bool ,默认 True
        排序
    - `ascending` : bool ,默认 False
        升序
    - `dropna` : bool ,默认 True
        不包括包含 NA 值的行数
